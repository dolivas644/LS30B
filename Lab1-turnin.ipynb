```python
#further exercises 1
# In our HPG model, the equation for H involved G . However, the exponent can be something other than 9. Call this exponent n and make an interactive that allows you to vary n between 2 and 12. At approximately what value of n does model start to exhibit persistent oscillations?
@interact
def HPG(n=(2,12,0.01)):
    var("H","P","G")
    t=srange(0,100,0.1)
    k1=0.2
    k2=0.21
    k3=0.18
    Pprime=H-k2*P
    Gprime=P-k3*G
    Hprime=(1/(1+ G^n)-(k1*H))
    sol1=desolve_odeint([Hprime,Pprime,Gprime], ics=[30,50,80], dvars=[H,P,G], times=t)
    p=list_plot(list(zip(t,sol1[:,0])), color="red", axes_labels=["time", "hormones"], legend_label=("pituary gland")) + list_plot(list(zip(t,sol1[:,1])), color="blue", axes_labels=["time", "hormones"], legend_label=("hypothalamus")) + list_plot(list(zip(t,sol1[:,2])), color="pink", axes_labels=["time", "hormones"], legend_label=("gonads"))
    show(p)
```

```python
#At approximate value of n 12 it starts to oscillate
```
