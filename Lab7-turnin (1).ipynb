```python
#Exercise 43
@interact
def interactive(d1=(-10, 10, .1), d2=(-10, 10, .1), target=vector([2,3])):
    u=vector([1, 2])
    v=vector([1, 0.5])
    p= d1*u + d2*v
    p1=plot(p, aspect_ratio=1, color="red") + plot(target, aspect_ratio=1, color="blue") + plot(u, aspect_ratio=1, color="green") + plot(v, aspect_ratio=1, color="yellow")
    show(p1)
```
