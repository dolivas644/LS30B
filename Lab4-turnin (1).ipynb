```python
#Exercise 5
@interact
def chaos_change(b1=(2,3)):
    var("X", "Y", "Z")
    a1 = 5
    a2=0.1
    b2= 2
    d1=0.4
    d2=0.01
    t= srange(0,1000,0.1)
    sol= desolve_odeint([X*(1-X) - ((a1*X)/(1+b1*X))*Y, ((a1*X)/(1+b1*X))*Y - (d1*Y) - ((a2*Y)/(1+ b2*Y))*Z, ((a2*Y)/(1+ b2*Y))*Z - d2*Z ],  ics = [1,0.5,8], dvars=[X,Y,Z], times= t)
    p=list_plot(list(zip(t,sol[:,0])), plotjoined=True, color="red", legend_label=["X:plants"])+list_plot(list(zip(t,sol[:,1])), plotjoined=True, color="brown", legend_label=["Y:herbivores"])+list_plot(list(zip(t,sol[:,2])), plotjoined=True, color="green", legend_label=["Z:carnivores"])
    show(p)
```
